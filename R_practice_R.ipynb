```r
#  data frames
df <- data.frame(name=c("Shom","Elizabeth"),title=c("King","Queen"))

df$name
df$title

df[1,] #first row
df[1,1] #first row first column

df[2,]  #second row

df[nrow(df)+1, ] <- c("x","position") #add to the end.

df[3,] #thrid row

paste(df, collapse=" ") #paste in one line
```

```r
#basic Function
f1 <- function() { print("Basic function")}
f1()

x <- readline("Please enter a name : ")
f2 <- function(x) { paste("Hello ",x)}
f2(x)

#fucntion with loop
x <- readline("Please enter a number : ")
f3 <- function(x) {
  for(i in 1:x) {
    print(paste("i equals to ", i))
    }
  }

f3(x)
```

```r
#lets try something a bit more complex
x <- as.numeric(readline("Please enter a number : ")) #
divideCheck <- function(x) {
  if (x %% 2 == 0) {
    print(paste(x, " is even."))
  }
  else if(x %% 3 == 0){
    print(paste(x, " is odd and divisble by 3."))
  }
  else{
    print(paste(x, " is odd."))
  }
}

divideCheck(x)
```
